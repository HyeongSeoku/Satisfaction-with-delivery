```python
df.columns=['날짜','시간','광역시도','시군구','읍면동','배달건수']
```

```python
df.groupby(['광역시도','시군구','읍면동','날짜','시간']).count().reset_index()
```

```python
# 각 광역시도, 시군구, 읍면동에서, 날짜를 컬럼을 2019-08-01부터 2021-08-31까지 차례로 도는데
# 만약 매치되는 날짜 컬럼이 없으면 해당 컬럼을 생성해주고, 있으면 그냥 패스

# 1. 일단 매치할 기준 날짜 리스트가 있어야 함 
entire_date = pd.date_range(start='2019-08-01', end='2021-09-01', freq='H')
entire_date = pd.DataFrame(entire_date)
entire_date
# -> 한 시간 단위까지 한번에 생성 되네!!!
```

```python
# 2. for 문으로 모든 지역을 돌아야하니까 일단 위치도 합쳐서 하나의 주소지로 만들자
df['지역'] = df['광역시도'] + '-' + df['시군구']+ '-' + df['읍면동']
df.drop(['광역시도', '시군구', '읍면동'], axis=1, inplace=True)
df
```

```python
# 3. 기준이 되는 전체달력시간리스트와 비교가 용이하도록 날짜와 시간도 합치기   : '2021-08-30 23:00:00'
# 시간 컬럼 데이터타입이 int64라 str으로 바꿔서 붙임
df['date_time'] = df['날짜']+ " " + df['시간'].apply(str) + ":00:00"
df.drop(['날짜','시간'], axis=1, inplace=True)
df
```

```python
# 4. 모든 개별 읍면동에 date_time이 누락된 거 없는지 기준셋entire_date랑 비교를 위해서는
# 모든 개별 읍면동을 for로 하나씩 돌아야 함으로, 일단 '지역'의 unique값으로 list 생성하기
all_areas = df['지역'].unique().tolist()
all_areas 
```

```python
df.지역.nunique()
```

all_areas : 모든 개별 지역('시도-시군구-읍면동')의 이름을 담고있는 list
entire_date : 우리가 커버하는 기간+시간 전체의 range 데이터프레임(배달건수 정보의 기준이 되는 개별 cell 단위) 

df['date_time']이랑 entire_date['date_time']이랑 포맷 일치하게 맞춰뒀음. 

for문 돌려서 all_area에서 하나씩 꺼내와서
해당 area랑 df['지역']이 일치하는 df을 걸러내서 date_time이랑 머지하면 됨

### 모든 고유 지역에는 18289개의 레코드가 있어야함

```python
df['date_time'] = pd.to_datetime(df['date_time'], format="%Y-%m-%d")
df
```

```python
entire_date.columns=['date_time']
entire_date
```

```python
# for문 돌려서 all_area에서 하나씩 꺼내와서 
# 해당 area랑 df['지역']이 일치하는 df을 걸러내서 date_time이랑 머지하면 됨

# for i in range(len(all_areas)):
#     area = all_areas[i]
#     df[df['지역']== area]
```

```python
# 일단 하나만 테스트로 해보자
# '경기도-고양시 일산동구-마두동'이랑 entire_date랑 머지해보자
test_area = df[df['지역']=='충청북도-충주시-칠금동']
# test_area
merge_test = pd.merge(test_area, entire_date, how='right', on='date_time')
merge_test['배달건수'] = merge_test['배달건수'].fillna(0)
merge_test['지역'] = merge_test['지역'].fillna(method='ffill')
merge_test['지역'] = merge_test['지역'].fillna(method='bfill')
merge_test
# merge_test[merge_test.지역.notnull()]
# merge_test[merge_test.지역.isnull()]
```

```python
merged_areas_df = merge_test
merged_areas_df
```

```python
for i in range(len(all_areas)):
    area = all_areas[i]
    merge_area = df[df['지역']==area]
    merged_area = pd.merge(merge_area, entire_date, how='outer', on='date_time')
    merged_area['배달건수'] = merged_area['배달건수'].fillna(0)
    merged_area['지역'] = merged_area['지역'].fillna(method='ffill')
    merged_area['지역'] = merged_area['지역'].fillna(method='bfill')
    merged_areas_df = merged_areas_df.append(merged_area)
merged_areas_df
```

```python
# 전체 레코드 갯수는 개별지역(913개) * 전체일수(762일)*시간대(24)로 나와야함 
# 마지막 2021-09-01 00:00:00시 하나가 더 들어가서 18289
18289*914 
#16697857
#16716146
# ffill로 안 채워지는 애들이 많이 있군...얘네는 bfill로 채워야하네
```

```python
df = merged_areas_df.drop_duplicates()
```
