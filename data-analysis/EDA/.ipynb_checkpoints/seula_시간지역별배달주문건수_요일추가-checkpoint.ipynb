```python
# 컬럼이름 넣어주기
df.columns=['날짜','시간대', '광역시도', '시군구', '읍면동', '배달건수']
df
```

```python
# 기간 확인
print(np.min(df['날짜']))
print(np.max(df['날짜']))
```

```python
# 날짜 데이터 타입 바꾸고 요일정보 추가하기
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y-%m-%d')

num_to_day = {
    0 : '월',
    1 : '화',
    2 : '수',
    3 : '목',
    4 : '금',
    5 : '토',
    6 : '일',
}

df['요일'] = [num_to_day[k] for k in df['날짜'].dt.dayofweek] ## 요일 칼럼
df.tail()
```
