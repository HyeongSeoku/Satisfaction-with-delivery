2019년 8월 ~ 2021년 8월까지 모든 공휴일 정보 가져오기

```python
holiday_cal = pd.DataFrame({
    'date':"",
    'day_name':"" ,
    'holiday': ""
    }, index=[0])
holiday_cal
```

```python
# 2019년 전체
url = 'http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getRestDeInfo?solYear=2019&numOfRows=30&ServiceKey=uDl6qNytS1TJ1zmnLP2wwqbSmk9W9hzyfQQdiX8a5zD2TdOAewteyUe4r9Nckeln8RCFEw11bz5JCFZ6OyHdGw%3D%3D'
result = requests.get(url)
soup = BeautifulSoup(result.text, "lxml")
items = soup.find_all("item")
for i in range(len(items)):
    holiday_cal=holiday_cal.append({'date':items[i].find('locdate').get_text(), 'day_name':items[i].find('datename').get_text() , 'holiday':items[i].find('isholiday').get_text()}, ignore_index=True)


# 2020년 붙이기
url = 'http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getRestDeInfo?solYear=2020&numOfRows=30&ServiceKey=uDl6qNytS1TJ1zmnLP2wwqbSmk9W9hzyfQQdiX8a5zD2TdOAewteyUe4r9Nckeln8RCFEw11bz5JCFZ6OyHdGw%3D%3D'
result = requests.get(url)
soup = BeautifulSoup(result.text, "lxml")
items = soup.find_all("item")
for i in range(len(items)):
    holiday_cal=holiday_cal.append({'date':items[i].find('locdate').get_text(), 'day_name':items[i].find('datename').get_text() , 'holiday':items[i].find('isholiday').get_text()}, ignore_index=True)

# 2021년 붙이기
url = 'http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getRestDeInfo?solYear=2021&numOfRows=30&ServiceKey=uDl6qNytS1TJ1zmnLP2wwqbSmk9W9hzyfQQdiX8a5zD2TdOAewteyUe4r9Nckeln8RCFEw11bz5JCFZ6OyHdGw%3D%3D'
result = requests.get(url)
soup = BeautifulSoup(result.text, "lxml")
items = soup.find_all("item")
for i in range(len(items)):
    holiday_cal=holiday_cal.append({'date':items[i].find('locdate').get_text(), 'day_name':items[i].find('datename').get_text() , 'holiday':items[i].find('isholiday').get_text()}, ignore_index=True)
holiday_cal

```
